```python
# here we are importing the libraries i.e json and time to read the data in the json format
import json
import time

# Initializing the variable
sale = " "
Discount = 0.10
IGST = 0.9
SGST = 0.9

#Loading the inventory record in the json
fd = open("Records.json", "r") # we are reding the data from json to string
js = fd.read()
fd.close()

record = json.loads(js) # we are converting string records to JSON

# Displaying the menu
print("---------------Menu----------------\n")
for keys in record.keys(): #.keys() is a method used to acess the keys of the dict
    print(keys,":", record[keys]["Name"],"\t|", record[keys]["Price"],"\t|", record[keys]["Quantity"])
print("-----------------------------------\n") 

# we are taking the required input from the user
ui_name = str(input("Entre the name        : "))
ui_mail = str(input("Entre the mail        : "))
ui_ph   = str(input("Entre the mob_num     : "))
ui_pr =   str(input("Entre the product ID  : "))
ui_quan = int(input("Entre the Quantity    : "))

# here we are checking that the quantity that user need whether we are having that much quantity or not.. if we have 
#.... that much quantity we will give it to the user and make his bill
if (record[ui_pr]['Quantity'] >= ui_quan):
   
    # here we are updating our quantity by decrementing it with the user quantity
    record[ui_pr]["Quantity"] -= ui_quan

    print("-----------------------------------\n")

    print("               Bill                \n")

    print("Name           :",record[ui_pr]["Name"])
    print("Price (RS)     :",record[ui_pr]["Price"])
    print("Quantity       :",ui_quan)
    print("IGST           :",IGST)
    print("SGST           :",SGST)

    print("------------------------------------")
    
    if record[ui_pr]["Price"] > 200:

        print("Billing amount :",record[ui_pr]["Price"] * ui_quan  - record[ui_pr]["Price"] * ui_quan * Discount * (IGST + SGST), "RS")
    
    else:
        print("Billing amount :",record[ui_pr]["Price"] * ui_quan  + record[ui_pr]["Price"] * ui_quan * (IGST + SGST), "RS")

    print("------------------------------------")
    # here we are upating our sales inventory with the details of the user / generating the sales in CSV format
    sale = "1"+","+ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(ui_quan)+","+str(record[ui_pr]["Price"])+","+str(ui_quan * record[ui_pr]["Price"])+","+time.ctime()+"\n"                                       
    
    
                                                 
# if we are not having enough quantity     
else:
    # here we are displaying the info about our quantity
    print("Sorry , we're not having enough quantity in our Inventory")
    print("We're only having", record[ui_pr]["Quantity"], "quantity")
    print("-----------------------------------")
    ch = input("Press Y/N to purchase: ")
    
    if ch == "YES" or ch == "yes": # if he agree to purchse all our remaining stock
    
        # again displaying the details
        print("-----------------------------------\n")    

        print("               Bill                \n")     

        print("Name           :",record[ui_pr]["Name"])
        print("Price (RS)     :",record[ui_pr]["Price"])     
        print("Quantity       :",record[ui_pr]["Quantity"])

        print("------------------------------------")
        
        if record[ui_pr]["Price"] > 200:

            print("Billing amount :",record[ui_pr]["Price"] * record[ui_pr]["Quantity"] - record[ui_pr]["Price"] * record[ui_pr]["Quantity"] * Discount * (IGST + SGST), "RS")
        
        else:
            print("Billing amount :",record[ui_pr]["Price"] * record[ui_pr]["Quantity"] + record[ui_pr]["Price"] * record[ui_pr]["Quantity"] * (IGST + SGST) , "RS")
        

        print("------------------------------------")
        # here we  are generating the sales with the quantity that we are having
        sales = "1"+","+ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(record[ui_pr]["Quantity"])+","+str(record[ui_pr]["Price"])+","+str(record[ui_pr]["Quantity"] * record[ui_pr]["Price"])+","+time.ctime()+"\n"                                     
        
        # here we are updating our inventory to 0 when all the quantity is purchased by the user
        record[ui_pr]["Quantity"] = 0
        
    
    #  if user dont want the quantity that we are hving we will print thanks
    else:
        if ch == "NO" or ch == "no":
            print("Thanks ! visit again")

    
# converting the dictionary to string
js = json.dumps(record)
#saving records  in json file
fd = open("Records.json", "w")
fd.write(js)
fd.close()

#saving records in sales file
fd = open("Sales.txt", "a")
fd.write(sale)
fd.close()

# at last we will print inventory updated
print("-------------------------------------------")
print("Thankyou for your order ,Inventory Updated!")
print("-------------------------------------------")


```
